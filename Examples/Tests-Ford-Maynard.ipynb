```python
def test_FM_I3(precision):    
    P = {nu < x1,
         x1 < x2,
         x2 < x3,
         x3 < 1/2,
         x1 + x2 + x3 == 1}
    I3a = sieve_integral(P, precision=precision, verbose=2, facteur=1.15)
    
    P = {nu < x1,
         x1 < x2,
         x2 < x3,
         x1 + x2 < (1-nu)/2,
         x1 + x2 + x3 == 1}
    I3b = sieve_integral(P, precision=precision, verbose=2, facteur=1.15)
    
    return -2*I3a-I3b
```

```python
I4_P = [{nu < x1,
         x1 < x2,
         x2 < x3,
         x3 < x4,
         x1 + x2 + x3 == 1,
         s < (1-nu)/2} for s in [x1+x2, x1+x3, x2+x3]]

def test_FM_I4(precision):
    return -add([sieve_integral(Q, precision=precision) for Q in I4_P])
```

```python
I5_base = [nu < x1,
           x1 < x2,
           x2 < x3,
           x3 < x4,
           x4 < x5,
           x1 + x2 + x3 + x4 + x5 == 1]

I5_P = [I5_base] + [I5_base + [s < (1-nu)/2] for s in (x3+x4, x1+x5, x2+x5, x3+x5, x4+x5)]

def test_FM_I5(precision):
    res = [sieve_integral(Q, precision=precision) for Q in I5_P]
    return res[0] - add(res[1:])

# This one is for I_5'
def test_FM_I5b(precision):
    tx = [x1, x2, x3, x4, x5]
    I5b_P = [I5_base + [tx[i] + tx[j] > 1/2] for i in range(5) for j in range(i+1, 5)]
    I5b = 0
    for Q in I5b_P:
        I5b += -2 * sieve_integral(Q, precision=precision)
    return I5b
```

```python
def chrono(commande):
    """
    Run 10 times the given command (as string), returns the computation times.
    """
    temps = []
    for _ in range(10):
        t1 = time()
        exec(commande)
        t2 = time()
        temps.append(t2 - t1)
    return temps
```
