```python
def polytope_from_bounds(L, var=0):
    """
    Here L is a lists [c1, c2, ..., c_k]
    where k is the number of variables in the to-be-computed integral (full-dimensional).
    Each c_i is a list [m, M], where m and M are themselves a list m = [m1, ... m_r],  M = [M_1, ..., M_s],
    corresponding to conditions  max(m) <= a_i <= min(M)
    """
    P = {add(a[j] for j in range(1, var+2)) == 1}
    for j in range(len(L)):
        for u in L[j][0]:
            P.add(a[j+1] > u)
        for u in L[j][1]:
            P.add(a[j+1] < u)
    return P



def add_polytope(L, var=0):
    global polytope_list
    if "polytope_list" not in globals():
        polytope_list = []
    polytope_list.append(polytope_from_bounds(L, var=var))


polytope_list = []
bounds_from_arxiv = []
```

```python
def compute_integral(Q, k, precision):
    return sieve_integral_harman(Q, (a[k], a[k-1]), precision=precision)
```

```python
value = {}
computation_time = {}
```

```python
def work(j, precision=20):
    
    Q = polytope_list[j]
    eqns, ieqs, v = symbolic_to_eqns(Q)
    P = Polyhedron(eqns = eqns, ieqs = ieqs, base_ring = QQ)
    k = P.ambient_dimension()
    
    print(f"polytope no. {j} ({P.ambient_dimension()} vars)")
    
    t1 = time()
    value[j] = K(compute_integral(Q, k, precision=precision))
    t2 = time()
    
    time_spent = round(t2 - t1, 2)
    
    print(f" I[{j}] ≤ {value[j]} versus {bounds_from_arxiv[j]} (time = {time_spent})")
    computation_time[j] = time_spent
    print(f"  difference {bounds_from_arxiv[j] - value[j].upper()}\n") 
```
