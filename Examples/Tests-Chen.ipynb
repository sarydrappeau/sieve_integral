## Integral C1

$$C_1 = I^*(P), \text{ where }  P = \{1/10 < t_1 < 1/3 < t_2 < t_3\}$$

```python
P0 = {t1 > 1/10,
      t1 < 1/3,
      t2 > 1/3,
      t3 > t2,
      t1 + t2 + t3 == 1}
```

```python
%%time
sieve_integral(P0, precision=40)
```

```python
temps = []
```

```python
for precision in range(14, 100):
    ts = timeit("sieve_integral(P0, precision=precision)")
    t = add(ts.series) / len(ts.series)
    n = (precision, t*1000)
    print(n)
    temps.append(n)
```

## Integral C2

$$ C_2 = I(P_1) - \tfrac12 I(P_2) - \tfrac12 I(P_3), $$
where
$$ P_1 = \{1 < t_3 < t_2-1 < t_1-2 < 2\}, $$
$$ P_2 = \{1 < t_3 < t_2-1 < 3 - 10t_1 < 2\}, $$
$$ P_3 = \{1 < t_3 < t_2-1 < 10 t_1-2 < 2\}. $$

```python
P1 = {1<t3,
      t3 < t2 - 1,
      t2-1 < t1 - 2,
      t1 - 2 < 2}
P2 = {1 < t3,
      t3 < t2-1,
      t2-1 < 3 - 10*t1,
      3 - 10*t1 < 2}
P3 = {1 < t3,
      t3 < t2-1,
      t2-1 < 10*t1 - 2,
      10*t1-2 < 2}
```

```python
%%time
v = [sieve_integral(Q, precision=40) for Q in [P1, P2, P3]]
v[0] - 1/2 * (v[1] + v[2])
```
